```python
import os
import shutil
import random

def split_dataset_by_fixed_numbers(input_dir, output_dir, val_count, test_count):
    """
    æ ¹æ“šæŒ‡å®šçš„é©—è­‰é›†å’Œæ¸¬è©¦é›†æ•¸é‡ä¾†åˆ†å‰²è³‡æ–™é›†ã€‚

    Args:
        input_dir (str): åŒ…å«é¡åˆ¥å­è³‡æ–™å¤¾çš„åŸå§‹è³‡æ–™é›†è·¯å¾‘ã€‚
        output_dir (str): è¦å„²å­˜åˆ†å‰²å¾Œè³‡æ–™é›†ï¼ˆtrain, validation, testï¼‰çš„ç›®æ¨™è·¯å¾‘ã€‚
        val_count (int): æ¯å€‹é¡åˆ¥è¦åˆ†é…åˆ°é©—è­‰é›†çš„åœ–ç‰‡æ•¸é‡ã€‚
        test_count (int): æ¯å€‹é¡åˆ¥è¦åˆ†é…åˆ°æ¸¬è©¦é›†çš„åœ–ç‰‡æ•¸é‡ã€‚
    """
    # ç¢ºä¿è¼¸å‡ºè³‡æ–™å¤¾å­˜åœ¨
    if os.path.exists(output_dir):
        # ç‚ºäº†é¿å…é‡è¤‡åŸ·è¡Œæ™‚ï¼ŒèˆŠè³‡æ–™å½±éŸ¿æ–°çµæœï¼Œå¯ä»¥é¸æ“‡æ€§åœ°åœ¨é–‹å§‹å‰åˆªé™¤èˆŠçš„è¼¸å‡ºè³‡æ–™å¤¾
        # shutil.rmtree(output_dir) 
        print(f"è¼¸å‡ºè³‡æ–™å¤¾ '{output_dir}' å·²å­˜åœ¨ã€‚å°‡åœ¨ç¾æœ‰çµæ§‹ä¸­æ–°å¢æˆ–è¦†è“‹æª”æ¡ˆã€‚")
    os.makedirs(output_dir, exist_ok=True)


    # å°æ¯å€‹é¡åˆ¥è³‡æ–™å¤¾é€²è¡Œè™•ç†
    for class_name in os.listdir(input_dir):
        class_path = os.path.join(input_dir, class_name)
        if not os.path.isdir(class_path):
            continue  # è·³ééè³‡æ–™å¤¾æª”æ¡ˆ

        # æ”¶é›†è©²é¡åˆ¥çš„æ‰€æœ‰åœ–ç‰‡
        images = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        random.shuffle(images)

        total = len(images)
        
        # æª¢æŸ¥åœ–ç‰‡ç¸½æ•¸æ˜¯å¦è¶³å¤ åˆ†é…
        if total < val_count + test_count:
            print(f"è­¦å‘Šï¼šé¡åˆ¥ [{class_name}] ç¸½åœ–ç‰‡æ•¸ ({total}) ä¸è¶³åˆ†é… {val_count} å¼µé©—è­‰é›†å’Œ {test_count} å¼µæ¸¬è©¦é›†ã€‚å°‡è·³éæ­¤é¡åˆ¥ã€‚")
            continue

        # æ ¹æ“šæŒ‡å®šæ•¸é‡é€²è¡Œåˆ†å‰²
        val_files = images[:val_count]
        test_files = images[val_count : val_count + test_count]
        train_files = images[val_count + test_count:]

        # è¤‡è£½åˆ°å„ç›®çš„è³‡æ–™å¤¾ï¼Œä¿æŒé¡åˆ¥çµæ§‹
        for file_list, split in [(train_files, 'train'), (val_files, 'validation'), (test_files, 'test')]:
            split_class_dir = os.path.join(output_dir, split, class_name)
            os.makedirs(split_class_dir, exist_ok=True)
            for filename in file_list:
                src = os.path.join(class_path, filename)
                dst = os.path.join(split_class_dir, filename)
                shutil.copy(src, dst)

        print(f"é¡åˆ¥ [{class_name}] åˆ†é…å®Œæˆï¼šTrain={len(train_files)}, Val={len(val_files)}, Test={len(test_files)}")

# ğŸ§ª ä½¿ç”¨ç¯„ä¾‹
input_folder = '../merged_data_from3_50to60'       # ä½ çš„åŸå§‹è³‡æ–™å¤¾
output_folder = '../dataset_full_en_aug7_56_new' # è¦å„²å­˜åˆ†å¥½é¡åˆ¥çš„è³‡æ–™å¤¾

# å‘¼å«ä¿®æ”¹å¾Œçš„å‡½å¼
split_dataset_by_fixed_numbers(input_folder, output_folder, val_count=120, test_count=20)
```

```python
import os
import shutil
import random

def split_dataset_by_class(input_dir, output_dir, train_ratio=0.7, val_ratio=0.2, test_ratio=0.1):
    # assert train_ratio + val_ratio + test_ratio == 1.0, "æ¯”ä¾‹ç¸½å’Œå¿…é ˆç‚º 1"

    # å°æ¯å€‹é¡åˆ¥è³‡æ–™å¤¾é€²è¡Œè™•ç†
    for class_name in os.listdir(input_dir):
        class_path = os.path.join(input_dir, class_name)
        if not os.path.isdir(class_path):
            continue  # è·³ééè³‡æ–™å¤¾

        # æ”¶é›†è©²é¡åˆ¥çš„æ‰€æœ‰åœ–ç‰‡
        images = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        random.shuffle(images)

        total = len(images)
        train_end = int(total * train_ratio)
        val_end = train_end + int(total * val_ratio)

        train_files = images[:train_end]
        val_files = images[train_end:val_end]
        test_files = images[val_end:]

        # è¤‡è£½åˆ°å„ç›®çš„è³‡æ–™å¤¾ï¼Œä¿æŒé¡åˆ¥çµæ§‹
        for file_list, split in [(train_files, 'train'), (val_files, 'validation'), (test_files, 'test')]:
            split_class_dir = os.path.join(output_dir, split, class_name)
            os.makedirs(split_class_dir, exist_ok=True)
            for filename in file_list:
                src = os.path.join(class_path, filename)
                dst = os.path.join(split_class_dir, filename)
                shutil.copy(src, dst)

        print(f"é¡åˆ¥ [{class_name}] åˆ†é…å®Œæˆï¼šTrain={len(train_files)}, Val={len(val_files)}, Test={len(test_files)}")

# ğŸ§ª ä½¿ç”¨ç¯„ä¾‹
input_folder = 'veg_img_eng_new'       # ä½ çš„åŸå§‹è³‡æ–™å¤¾
output_folder = 'dataset_full_en_aug5(no mix up)'       # è¦å„²å­˜åˆ†å¥½é¡åˆ¥çš„è³‡æ–™å¤¾

split_dataset_by_class(input_folder, output_folder)
```
